```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
```

```python
def generate_lifetime(a, b, size):
    return np.random.gamma(shape=a, scale=b, size=size)

def measure_lifetime(a, b, num_sam):
    lifetime = generate_lifetime(a, b, num_sam)
    plt.hist(lifetime, bins=30, density=True, alpha=0.8)
    plt.show()
    
measure_lifetime(2.5, 1.2, 1000)
```
