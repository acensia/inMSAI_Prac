```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
lambda_param = 3

num_sim = 1000
e1 = np.random.poisson(lambda_param, 500)
e2 = np.random.poisson(lambda_param, 1000)
e3 = np.random.poisson(lambda_param, 2000)
e4 = np.random.poisson(lambda_param, 3000)

plt.hist(e4, bins=10, alpha=0.8, color="r")
plt.hist(e3, bins=10, alpha=0.8, color="g")
plt.hist(e2, bins=10, alpha=0.8, color="b")
plt.hist(e1, bins=10, alpha=0.8, color="k")
plt.show()
```
