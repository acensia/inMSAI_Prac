```python
import numpy as np
import soundfile as sf

import IPython.display as ipd
```

```python
file_path = "./blues.00001.wav"
data, sr = sf.read(file_path)

ipd.Audio(data, rate=sr)
```

Create Noize

```python
noise_amp = 0.5*np.random.uniform()*np.amax(data)
noise = noise_amp * np.random.normal(size=data.shape)

print(noise_amp)
print(noise)

data_noise = data.astype("float32") + noise.astype("float32")
```

```python
save_path = "./sample_noise_add.wav"
sf.write(save_path, data_noise, sr)
```

```python
sample_wav = "./sample_noise_add.wav"
sample_data, sample_sr = sf.read(sample_wav)

ipd.Audio(sample_data, rate=sample_sr)
```

Compare noised audio with graph

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.plot(data)
plt.title('Original')
plt.subplot(1, 2, 2)
plt.plot(sample_data)
plt.title('Noise Add')
plt.show()
```

# 05
Data Augmentation for sound data

```python
#np, ply, librosa, sf
import librosa

data, sr = librosa.load('./blues.00001.wav')

data_stretch = librosa.effects.time_stretch(data, rate=0.8)
```

```python
save_path = "./sample_stretch.wav"  
sf.write(save_path, data_stretch, sr)
```

Compare stretched data with original

```python
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
librosa.display.waveshow(data, sr=sr, alpha=0.8)

plt.subplot(1, 2, 2)
librosa.display.waveshow(data_stretch, sr=sr, alpha=0.8)

plt.title('Stretch')
plt.show()
```
