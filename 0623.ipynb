Binary Cross Entropy Loss

```python
import numpy as np
```

```python
def binary_cross_entropy_loss(y_true_tmp, y_pred_tmp,):
    epsilon = 1e-7
    n = y_true_tmp.shape[0]
    print("n val >> ", n)
    loss = -np.mean(y_true_tmp * np.log(y_pred_tmp + epsilon) + (1-y_true_tmp)*np.log(1-y_pred_tmp + epsilon))
    return loss


y_true_tmp = np.array([1, 0, 1, 1])
y_pred_tmp = np.array([0.99, 0.1, 0.8, 0.7])

loss = binary_cross_entropy_loss(y_true_tmp, y_pred_tmp)
print(loss)
```

```python
def cross_entropy_loss(y_true, y_pred, c):
    n_tmp = y_true.shape[0]
    print("n_value >> ", n_tmp)
    ce_loss = -1 / n_tmp * np.sum(y_true * np.log(y_pred))
    
    return ce_loss

y_true01 = np.array([[1, 0, 0], [0, 1 ,0], [0, 0, 1]])
y_pred01 = np.array([[0.7, 0.2, 0.1], [0.1, 0.8, 0.1], [0.2, 0.3, 0.5]])
c=3

ce_loss_val = cross_entropy_loss(y_true01, y_pred01, c)
print(ce_loss_val)
```
