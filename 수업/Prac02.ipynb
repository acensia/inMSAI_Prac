```python
import pandas as pd
import random
import os

dates = pd.date_range(start="2021-01-04", periods=100, freq='D')
samsung_prices = [random.randint(80000,90000) for _ in range(100)]
samsung_data = {'Date' : dates, '005930.KS' : samsung_prices}
samsung_df = pd.DataFrame(samsung_data)
samsung_df.set_index("Date", inplace=True)

lg_prices = [random.randint(140000,160000) for _ in range(100)]
lg_data = {'Date' : dates, '066570.KS' : lg_prices}
lg_df = pd.DataFrame(lg_data)
lg_df.set_index("Date", inplace=True)

df = pd.concat([samsung_df, lg_df], axis=1)
df = df.loc[:, ['005930.KS', '066570.KS']]
df.columns = ['Samsung', 'LG']

print(df)

os.makedirs("./data", exist_ok=True)
df.to_csv("./data/stock_exam.csv")
```

```python
import numpy as np

np.random.seed(42)
height = np.random.normal(loc=170, scale=10, size=100)
weight = np.random.normal(loc=70, scale=5, size=100)

print(height)
print(weight)

data = pd.DataFrame({'Height' : height, 'Weight' : weight})
data.to_csv("./data/human_data.csv", index=False)
```

```python
h_mean, h_std = norm.fit(h_data)
h_x = np.linspace(h_data.min(), h_data.max(), 100)
h_y = norm.pdf(h_x, h_mean, h_std)

w_mean, w_std = norm.fit(w_data)
w_x = np.linspace(w_data.min(), w_data.max(), 100)
w_y = norm.pdf(w_x, w_mean, w_std)
```

```python

plt.hist(height, bins=10, density=True, alpha=0.5, label = "Height")
plt.plot(h_x, h_y, label="Height Distribution")
plt.hist(weight, bins=10, density=True, alpha=0.5, label = "Weight")
plt.plot(w_x, w_y, label="Weight Distribution")
plt.legend(loc="upper right")
plt.show()
```
