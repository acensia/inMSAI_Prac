Create random time series data and visualize upward trend

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.rand(100).cumsum()

data = pd.DataFrame({
    'Date' : dates,
    'Values' : values
})

def trend_show(n):
    plt.plot(data['Date'], data[n], label=n)
    plt.legend()
    plt.xticks(rotation=-45)
    plt.show()
```

```python
data['Upward_Trend'] = np.linspace(0, 50, len(data))
print(data)
trend_show("Upward_Trend")
```

```python
data['Downward_Trend'] = np.linspace(50, 0, len(data))
trend_show("Downward_Trend")
data["Flat_Trend"] = np.full(len(data), 30)
trend_show("Flat_Trend")
```

# Time series Data search & visualize

```python
# Create data and DataFrame
size = 1000
dates = pd.date_range(start="2023-01-01", periods=size, freq="D")
values = np.random.randn(size)

Data = pd.DataFrame({
    "Dates" : dates,
    "Values":values
})

# Visualize
plt.plot(Data["Dates"], Data["Values"])
plt.xticks(rotation=-45)
plt.show()

plt.scatter(Data["Dates"], Data["Values"])
plt.xticks(rotation=-45)
plt.show()

plt.bar(Data["Dates"], Data["Values"])
plt.xticks(rotation=-45)
plt.show()

plt.boxplot(Data["Values"])
plt.show()
```

```python
import seaborn as sns

pivot_table = Data.pivot(index='Dates', columns='Values', values='Values')

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='Reds', cbar=True)
plt.show()
```
