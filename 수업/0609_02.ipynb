```python
import cv2
import matplotlib.pyplot as plt
```

```python
sample_img = cv2.imread("./data/ex_image.png")
sample_logo = cv2.imread("./data/ex_image_logo.png")

sample_img = cv2.cvtColor(sample_img, cv2.COLOR_BGR2RGB)
sample_logo = cv2.cvtColor(sample_logo, cv2.COLOR_BGR2RGB)

plt.imshow(sample_img)
plt.show()

plt.imshow(sample_logo)
plt.show()
```

```python
blended = cv2.addWeighted(sample_img, 0.7, sample_logo, 0.3, 0)

plt.imshow(blended)
plt.show()
```

```python
import numpy as np

img_rect = np.ones((400, 400), dtype="uint8")
cv2.rectangle(img_rect, (50, 50), (300, 300), (255,255,255), -1)

img_circle = np.ones((400,400), dtype="uint8")
cv2.circle(img_circle, (300, 300), 50, (255,255,255), -1)

plt.imshow(img_rect, 'gray')
plt.show()
plt.imshow(img_circle, 'gray')
plt.show()
```

with A, B data and, or, xor, not

```python
bwAnd = cv2.bitwise_and(img_rect, img_circle)
plt.imshow(bwAnd, 'gray')
plt.show()

bwOr = cv2.bitwise_or(img_rect, img_circle)
plt.imshow(bwOr, 'gray')
plt.show()

bwXor = cv2.bitwise_xor(img_rect, img_circle)
plt.imshow(bwXor, 'gray')
plt.show()

bwNot = cv2.bitwise_not(img_rect, img_circle)
plt.imshow(bwNot, 'gray')
plt.show()
```

```python
img_rect_temp = np.ones((500, 1000), dtype='uint8')

mask = np.zeros(img_rect_temp.shape, dtype='uint8')

cv2.rectangle(mask, (60, 50), (280, 280), (255, 255, 255), -1)
cv2.rectangle(mask, (420, 50), (550, 230), (255, 255, 255), -1)
cv2.rectangle(mask, (750, 50), (920, 280), (255, 255, 255), -1)

plt.imshow(mask,'gray')
plt.show()
```
