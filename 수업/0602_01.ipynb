```python
import numpy as np
from sklearn.decomposition import FactorAnalysis

x = np.random.randn(100, 5)
# print(x)
```

```python
n_components = 2
factorAnal = FactorAnalysis(n_components=n_components, random_state=0)
x_factor = factorAnal.fit_transform(x)

print("size of orginal dataset",x.shape)
print(f"factor analysis result - factor coef : {n_components}")
print(x_factor)
```
