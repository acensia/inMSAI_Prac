```python
img_path = "./data/24.jfif"
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img_blur = cv2.blur(img, (7,7))
plt.imshow(img)
plt.show()
plt.imshow(img_blur)
plt.show()
```

```python
# gaussian blur
img_g = cv2.GaussianBlur(img, (9, 9), 0)
plt.imshow(img_g)
plt.show()
```

```python
kernel_ = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]
                    ])

img_sharp = cv2.filter2D(img, -1, kernel_)

plt.imshow(img)
plt.show()


plt.imshow(img_sharp)
plt.show()
```
