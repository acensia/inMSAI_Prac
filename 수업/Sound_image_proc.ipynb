```python
import os
import glob
from tqdm import tqdm
from PIL import Image

# if linux
# slash = '/'
# if window
slash = '\\'
```

```python
def process_img_resize(pil_img, background_color):
    w, h = pil_img.size
    
    if w == h :
        return pil_img
    elif w > h :
        img = Image.new(pil_img.mode, (w, w), background_color)
        img.paste(pil_img,(0,(w - h)//2))
        return img
    else :
        img = Image.new(pil_img.mode, (h, h), background_color)
        img.paste(pil_img,(0,(h - w)//2))
        return img
    
    
def resize_with_padding(pil_img, new_size, background_color):
    img = process_img_resize(pil_img, background_color)
    img = img.resize(new_size, Image.ANTIALIAS)
    
    return img
        
```
