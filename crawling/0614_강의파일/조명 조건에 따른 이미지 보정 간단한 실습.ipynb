######  1. OpenCV 라이브러리를 활용하여 이미지의 밝기를 조정 간단한 실습

cv2.pow()

- 이미지의 각 픽셀 값에 대해 지수 연산을 수행하는 함수입니다. 

- 이미지의 밝기를 조절하는 경우에는 cv2.pow() 함수를 사용하여 이미지의 각 픽셀 값에 대해 지수 연산을 수행합니다. 이때, power 값이 1보다 작은 경우에는 이미지가 어두워지고, 1보다 큰 경우에는 이미지가 밝아집니다. 

```python
import cv2
import matplotlib.pyplot as plt

# 이미지 읽기 
image = cv2.imread("./image01.jpeg")

# BGR -> RGB 변경 
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 이미지 밝기 조정 
gamma = 2.5
img_corrected = cv2.pow(image/255.0, gamma)
img_corrected = img_corrected*255.0

# 원본 이미지 밝기 조정 결과 모습
plt.imshow(image)
plt.title("org image")
plt.show()

# 이미지 밝기 조정 결과 모습 
plt.imshow(img_corrected)
plt.title("img_corrected")
plt.show()
```
