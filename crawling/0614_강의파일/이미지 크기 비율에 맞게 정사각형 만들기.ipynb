#### 이미지 크기 비율에 맞게 정사각형 이미지 만들기 

```python
import matplotlib.pylab as plt
from PIL import Image

def expend2square(pil_img, background_color) : 
    width, heigth = pil_img.size
    
    if width == heigth :
        return pil_img
    
    elif width > heigth :
        result = Image.new(pil_img.mode, (width, width), background_color)
        result.paste(pil_img, (0, (width - heigth) // 2))
        return result
    else :
        result = Image.new(pil_img.mode, (heigth, heigth), background_color)
        result.paste(pil_img, ((heigth - width) //2,0))
        return result 
```

```python
def resize_with_padding(pil_img, new_size, background_color) :
    img = expend2square(pil_img, background_color)
    img = img.resize((new_size[0], new_size[1]), Image.ANTIALIAS)
    
    return img
```

```python
img = Image.open("./image01.jpeg")
img_new = resize_with_padding(img, (300,300), (0,0,255))

plt.imshow(img)
plt.show()

plt.imshow(img_new)
plt.show()
```
