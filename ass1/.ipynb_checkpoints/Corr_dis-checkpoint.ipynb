```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
```

```python
def create_score(g, s, sub):
    score = np.random.normal(loc=75, scale=10, size=g*s).astype(int)
    score = np.clip(score, 0, 100)
    
    return pd.DataFrame({sub : score})
```

```python
grade = 3
students = 100
math_= create_score(grade, students, 'math')
eng_= create_score(grade, students, 'eng')
kor_= create_score(grade, students, 'kor')
grades = pd.DataFrame({'Grade':[int(i/students +1) for i in range(grade*students)]})
Scores = pd.concat([grades, math_, eng_, kor_], axis = 1)
Scores.set_index('Grade', inplace =True)

try:
    os.makedirs('./data', exist_ok=False)
except:
    pass
Scores.to_csv(''./data/student_scores.csv')
```

```python
Scores = pd.read_csv('./data/student_scores.csv')
print(Scores)
```

```python
corr_matrix = Scores.corr()
sns.heatmap(corr_matrix, annot = True, cmap = 'Reds')
plt.title('corr')
plt.show()
```

```python
print(Scores['math'].describe())
print('/////////////////////////////////////')
print(Scores['eng'].describe())
print('/////////////////////////////////////')
print(Scores['kor'].describe())
```
